**Practice Problem 6**

```python
import pandas as pd
from itertools import combinations
from docplex.mp.model import Model

# 1: Define distance matrix
df_distance = pd.DataFrame(
    [
        [0, 5, 6, 8],
        [5, 0, 13, 8],
        [6, 13, 0, 9],
        [8, 8, 9, 0]
    ],
    columns=[0, 1, 2, 3],
    index=[0, 1, 2, 3]
)

list_nodes = list(df_distance.index)  # save the nodes as a list
list_edges = [(i, j) for i in list_nodes for j in list_nodes if i < j]  # save edges as list

# 2: Define the model and decision variable x
mdl = Model(name="STSP_DFJ_CPLEX")  # create a CPLEX model
var_x = mdl.binary_var_dict(list_edges, name="x")

# 3: Define the constraints

# 3-1: Degree-2 constraint — each node has 2 incident edges
for j in list_nodes:
    incident = [e for e in list_edges if j in e]  # edges involving node j
    mdl.add_constraint(mdl.sum(var_x[i, j] for (i, j) in incident) == 2)

# 3-2: Subtour elimination — DFJ constraints

# 3-2-1: Create list of subsets (size 2 to n-1)
subsets = []
n = len(list_nodes)
for k in range(2, n):
    subsets.extend(combinations(list_nodes, k))

# 3-2-2: Add DFJ constraints based on each subset
for S in subsets:
    S = set(S)
    cut_edges = [(i, j) for (i, j) in list_edges if (i in S and j not in S) or (j in S and i not in S)]
    mdl.add_constraint(mdl.sum(var_x[i, j] for (i, j) in cut_edges) >= 2)

# 4: Objective — minimize total distance
mdl.minimize(
    mdl.sum(df_distance.loc[i, j] * var_x[i, j] for (i, j) in list_edges)
)

# 5: Solve and print
solution = mdl.solve(log_output=True)

if solution:
    print("Objective value (DFJ):", mdl.objective_value)
    print("Selected edges:")
    for (i, j) in list_edges:
        if var_x[i, j].solution_value > 0.5:
            print(f"{i} -- {j}")
else:
    print("No solution found.")

# Optional: Show the model formulation
print("\nMODEL FORMULATION:")
print(mdl.export_as_lp_string())
```

```python
# 1. Import required libraries and the city data
import numpy
import pandas as pd
from geopy.distance import geodesic
from itertools import combinations
from docplex.mp.model import Model

city_df = pd.read_csv('lab_2_data.csv')

# 2. Calculate the distance matrix using WGS-84 system
#create a list of city coordinates
city_coords = list(zip(city_df['Latitude'], city_df['Longitude']))
city_names = city_df['City Name'].tolist()
n_cities = len(city_names)


distance_matrix = pd.DataFrame(0, index=city_names, columns=city_names)


# Fill the distance matrix, only upper triangle, as STSP is symmetric
for i, name_i in enumerate(city_names):
    for j, name_j in enumerate(city_names):
        if i < j:
            # Calculate geodesic distance and round to integer
            dist = int(geodesic(city_coords[i], city_coords[j]).km)
            distance_matrix.loc[name_i, name_j] = dist
            distance_matrix.loc[name_j, name_i] = dist  # Symmetric



# 3. Solve the STSP problem using CPLEX 

nodes = city_names
edges = [(i, j) for i in nodes for j in nodes if i < j]

mdl = Model(name="STSP_DFJ_CPLEX")

# Decision variables: x[i, j] = 1 if edge (i, j) is in the tour
x = mdl.binary_var_dict(edges, name="x")


# Degree-2 constraint: Each node has exactly 2 incident edges
for node in nodes:
    incident_edges = [e for e in edges if node in e]
    mdl.add_constraint(mdl.sum(x[e] for e in incident_edges) == 2)

# Subtour elimination (DFJ): For every proper subset S, at least 2 edges leave S
from itertools import chain, combinations as it_combinations

def all_subsets(iterable):
    "Returns all non-empty, non-full subsets"
    s = list(iterable)
    return chain.from_iterable(it_combinations(s, r) for r in range(2, len(s)))

for S in all_subsets(nodes):
    S = set(S)
    if len(S) < n_cities and len(S) > 1:
        cut_edges = [(i, j) for (i, j) in edges if (i in S and j not in S) or (j in S and i not in S)]
        mdl.add_constraint(mdl.sum(x[e] for e in cut_edges) >= 2)

# Objective: Minimize total tour distance
mdl.minimize(mdl.sum(distance_matrix.loc[i, j] * x[i, j] for (i, j) in edges))

# Solve the model
solution = mdl.solve(log_output=True)

# Output the results
if solution:
    print("Objective value - Total Distance:", int(mdl.objective_value))
    print("Selected edges in the optimal tour:")
    for (i, j) in edges:
        if x[i, j].solution_value > 0.5:
            print(f"{i} -- {j}")
    # Build adjacency dictionary from selected edges
    from collections import defaultdict, deque

    adj = defaultdict(list)
    for (i, j) in edges:
        if x[i, j].solution_value > 0.5:
            adj[i].append(j)
            adj[j].append(i)

    # Reconstruct the tour in order
    tour = []
    visited = set()
    current = nodes[0]  # start from the first city
    prev = None
    for _ in range(len(nodes)):
        tour.append(current)
        visited.add(current)
        # Go to the next unvisited neighbor
        neighbors = [n for n in adj[current] if n != prev]
        if neighbors:
            prev, current = current, neighbors[0]
        else:
            break
    tour.append(tour[0])  # complete the cycle

    print("\nOrdered tour:")
    print(" -> ".join(tour))


```
